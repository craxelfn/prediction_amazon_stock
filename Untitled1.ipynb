```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
from datetime import timedelta

# Load the dataset
data = pd.read_csv('amazon_stock_data.csv')

# Check for any missing values
print(data.isnull().sum())

# Fill missing values (if any) or drop rows with missing data
data['Close'].fillna(method='ffill', inplace=True)

# Convert the 'Date' column to datetime and set it as the index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Ensure the 'Close' column is numeric
data['Close'] = pd.to_numeric(data['Close'], errors='coerce')

# Drop rows with any NaN values after conversion
data.dropna(inplace=True)

# Prepare the data
# Use 'Close' prices for prediction
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1, 1))

# Prepare the dataset for the model (assuming it was trained on 60 days of data)
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 60
X, y = create_dataset(scaled_data, time_step)

# Reshape the data to be compatible with the model
X = X.reshape(X.shape[0], X.shape[1], 1)

# Load the pre-trained model
model = load_model('best_model.h5')

# Make predictions for the next 10 days
last_60_days = scaled_data[-time_step:].reshape(1, time_step, 1)  # Ensure the shape is [1, time_step, 1]
predicted_prices = []

for _ in range(10):
    pred = model.predict(last_60_days)
    predicted_prices.append(pred)
    # Update last_60_days by appending the prediction
    last_60_days = np.append(last_60_days[:, 1:, :], pred.reshape(1, 1, 1), axis=1)

# Inverse transform the predictions back to original scale
predicted_prices = scaler.inverse_transform(np.array(predicted_prices).reshape(-1, 1))

# Create dates for the next 10 days
last_date = data.index[-1]
predicted_dates = [last_date + timedelta(days=i) for i in range(1, 11)]

# Prepare the plot
plt.figure(figsize=(10,6))

# Plot the original stock prices
plt.plot(data.index, data['Close'], label='Actual Stock Price')

# Plot the predicted stock prices
plt.plot(predicted_dates, predicted_prices, label='Predicted Stock Price', color='red')

# Labels and Title
plt.title('Amazon Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend()

# Show the plot
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
